```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Training losses

### U-net pretraining
the U-net generator was pretrained on a simple l1 loss for 30 epochs


```python
# Read the CSV files
training_loss_df = pd.read_csv('../training_logs/unet_pretraining/unet_train_loss.csv')
validation_loss_df = pd.read_csv('../training_logs/unet_pretraining/unet_valid_loss.csv')

# Extract data
train_epochs = training_loss_df['Epoch'].values
train_loss = training_loss_df['Train L1 Loss'].values
val_epochs = validation_loss_df['Epoch'].values
val_loss = validation_loss_df['Validation L1 Loss'].values

# Plot training loss
plt.figure(figsize=(8, 4))
plt.plot(train_epochs, train_loss, label='Training Loss (L1)', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('U-Net Training Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot validation loss
plt.figure(figsize=(8, 4))
plt.plot(val_epochs, val_loss, label='Validation Loss (L1)', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('U-Net Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot both losses on the same graph
plt.figure(figsize=(8, 4))
plt.plot(train_epochs, train_loss, label='Training Loss (L1)', color='blue')
plt.plot(val_epochs, val_loss, label='Validation Loss (L1)', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('U-Net Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.savefig(f'losses/unet_pretraining.png')
plt.show()
```

### Training 'vanilla' GAN using a pretrained(for 30 epochs) Unet as a generator

 This GAN was trained using BCE (Binary Cross entropy) Loss for additional 20 epochs

```python
df = pd.read_csv('../training_logs/vanilla/training_loss.csv')
epochs = df['Epoch'].values

loss_D = df['loss_D'].values
loss_D_real = df['loss_D_real'].values
loss_D_fake = df['loss_D_fake'].values

loss_G = df['loss_G'].values
loss_G_gan= df['loss_G_GAN'].values
loss_G_L1 = df['loss_G_L1'].values

# plotting Generator loss
plt.figure(figsize=(8, 4))
plt.plot(epochs, loss_G, label='Generator Total Loss', color='orange')
plt.plot(epochs, loss_G_gan, label='Generator Adversarial Loss', color='green')
plt.plot(epochs, loss_G_L1, label='Generator L1 Loss', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Generator loss, Vanilla GAN')
plt.legend()
plt.grid(True)
plt.savefig(f'losses/vanilla_generator.png')
plt.show()

# plotting Discriminator loss
plt.figure(figsize=(8, 4))
plt.plot(epochs, loss_D, label='Discriminator Total Loss', color='orange')
plt.plot(epochs, loss_D_real, label='Discriminator Real Loss', color='green')
plt.plot(epochs, loss_D_fake, label='Discriminator Fake Loss', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Discriminator loss, Vanilla GAN')
plt.legend()
plt.grid(True)
plt.savefig(f'losses/vanilla_discriminator.png')
plt.show()

# plotting Adversarial loss
plt.figure(figsize=(8, 4))
plt.plot(epochs, loss_D, label='Discriminator Loss', color='red')
plt.plot(epochs, loss_G_gan, label='Generator Loss', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Adversarial loss, Vanilla GAN')
plt.legend()
plt.grid(True)
plt.savefig(f'losses/vanilla_adversarial.png')
plt.show()
```

## Evaluating 5 models on 6 metrics

```python
df = pd.read_csv("../metrics_coco_valid_2k.csv")
sns.set_theme(style="whitegrid", palette="Set2")

metric_names = df['Metric Name'].unique()

for metric in metric_names:
    metric_df = df[df['Metric Name'] == metric]

    plt.figure(figsize=(8, 5))
    
    sns.barplot(x="Model Name", y="Metric Value", data=metric_df, palette="Set1", hue="Model Name", legend=False)

    for p in plt.gca().patches:
        plt.text(p.get_x() + p.get_width() / 2., p.get_height(),
                 f'{p.get_height():.5f}',  # Format the number to 5 decimals
                 ha='center', va='center', fontsize=14, color='black', fontweight='bold')

    plt.title(f'Comparison of {metric} Across Models', fontsize=16, fontweight='bold')
    plt.xlabel('Model Name', fontsize=14, fontweight='bold')
    plt.ylabel('Metric Value', fontsize=14, fontweight='bold')
  
    plt.xticks(fontsize=12, fontweight='bold', ha='center', family='Arial', rotation=15)
    plt.yticks(fontsize=12)

    plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()

    plt.savefig(f'metrics/{metric}.png')
    plt.show()
```
